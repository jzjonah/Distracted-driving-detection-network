```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import numpy as np
import cv2
#path holding pictures on non distracted drivers
path_root ='/content/drive/MyDrive/train'
path1 =path_root+'/c0/*.*'
#set up arrays to hold images for training and testing
images = np.zeros((3993,120,160))
images_test = np.zeros((3491,120,160))

#set up arrays for holding the labels of the images
labels2= []
labels_test2= []
#set up counters for keeping track of amount of images in the arrays
counter = 0
counter_test= 0

#loops through every image in path holding non distracted driving
for file in glob.glob(path1):
#first 1500 images labels added to training array rest for testing
  if counter < 1500:
    labels2.append(1)
  else:
    labels_test2.append(1)
  #Read image and reshape to a 2D array
  img = cv2.imread(file,cv2.IMREAD_GRAYSCALE)
  img= np.reshape(img,(480,640))
  #shrink image for faster training and less storage
  img =cv2.resize(img,(160,120))
  #change data type to float32 and reduce values to between 0-1
  img = img.astype('float32')/255
  #save first 1500 images to training array and rest to testing
  if counter < 1500:
    images[counter] = img
    counter = counter + 1
  else:
    images_test[counter_test] =img
    counter_test = counter_test +1
```

```python
import struct
import numpy as np
import keras

from keras import metrics
from keras.layers import Conv2D
from keras.layers import Add
from keras.layers import Flatten
from keras.layers import Input
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import AveragePooling2D
from keras.layers import MaxPooling2D
from keras.layers import BatchNormalization
from keras.layers import LeakyReLU
from keras.layers import ZeroPadding2D
from keras.layers import UpSampling2D
from keras.activations import linear
from keras.layers import Softmax
from keras.layers import Activation
from keras.layers.merge import add, concatenate
from keras.models import Model

#determine number of classes for the data
n_values = np.max(labels2) + 1
#convert labels to one hot encoding
y_train2 = np.eye(n_values)[labels2]
model = keras.models.Sequential()

model.add(Conv2D(256, (3, 3), input_shape=(120,160,1), activation='relu', padding='valid'))
model.add(MaxPooling2D(pool_size=(2, 2), padding='valid'))
model.add(Conv2D(256, (3, 3), activation='relu', padding='valid'))
model.add(MaxPooling2D(pool_size=(2, 2), padding='valid'))
model.add(Flatten()) 
model.add(Dense(100, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(10, activation='softmax'))

#compile model 
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=[
        metrics.Precision(),
        metrics.Recall(),
        'accuracy'
        
    ])
print(model.summary())
print("Fit model on training data")
#reshape data to be able to propperly run in the network 
images = np.reshape(images,(3993,120,160,1))
#train model usings the images as the input and y_train2 as expected output
history = model.fit(
    images,
    y_train2,
    batch_size=32,
    epochs=3,
    
)

model.save('/content/drive/MyDrive/model/distracted_driving.h5')
```
